```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
# Loading the dataset
data=pd.read_csv("wine.csv")
data
```

 <b>*We can see by default the data is divided into 3 types(or clusters) in the 1st column*</b>

```python
# Droping type column
wine=data.drop('Type',axis=1)
wine
```

```python
wine.info()
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
w_norm = norm_func(wine)
w_norm.head()
```

# PCA

```python
# Preforming PCA with 3 principal components
pca = PCA(3)
pca_values = pca.fit_transform(w_norm)
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color="blue")
```

```python
 pca_values
```

```python
# Converting the PCA values to dataframe
```

```python
pca_df=pd.DataFrame(pca_values,columns=['PC1','PC2','PC3'])
pca_df
```

## Checking weather the pca values give same no.of clusters by k means and Hierarchical Clustering

### Hierarchical Clustering 

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
dendrogram = sch.dendrogram(sch.linkage(pca_values, method='ward'))
```

<b>Here we can see the our pca_values are clearly divided into 3 clusters same as the our default data is also divided into 3 types</b>

```python
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(pca_values)
y_hc
```

```python
pca_df['h_clusterid'] = hc.labels_
```

### k means clustering

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(pca_values)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

<b>
1.From above Elbow diagram we can see the curve has steep bend at 3,so we can consider that optimum clusters as 3
    
    
2.so by k means also we got as 3 clusters as default data</b>

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=42)
clusters_new.fit(pca_values)
```

```python
clusters_new.labels_
```

```python
pca_df['kmeans_cluster'] = clusters_new.labels_
```

```python
pca_df['type']=data.iloc[:,0]
```
